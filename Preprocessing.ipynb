```python

import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\BIA PROJECT\Telco_Customer_Churn.csv")
data.head()
```

# Data Preproccessing

```python
# Setting 
pd.set_option("display.max_columns",None)
```

```python
# Creating Backup
df=data.copy()
df.head()
```

```python
df.info()
```

```python
# Duplicate
print("Total duplicate rows:")
print(df.duplicated().sum())
df[df.duplicated()]
```

```python
# Drop CustomerID
df.drop(columns='customerID',axis=1,inplace=True)
```

```python
# Change datatype of "total CHarges"
df['TotalCharges']= pd.to_numeric(df['TotalCharges'], errors ='coerce')
df['TotalCharges'].dtype
```

```python
# Drop NUll
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
# Save Clean Data
df.to_csv("data_cleaned.csv",index=False)
```
